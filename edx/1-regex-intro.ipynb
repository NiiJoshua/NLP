```python
import re
```

```python
pattern = r"floor|floors"
print(re.findall(pattern,"my block has 15 floors, I live on floor 5"))
```

```python
# Greedy evaluation
expression = "Plants (e.g., trees) and animals (e.g., mamals) make up the biosphere."
pattern = r"\(.*\)"
print(re.findall(pattern, expression))
```

```python
# Lazy evaluation - the inclusion of "?"
expression = "Plants (e.g., trees) and animals (e.g., mamals) make up the biosphere."
pattern = r"\(.*?\)"
print(re.findall(pattern, expression))
```

```python
sentence = "Only 8 of out 46 customers paid the full price of 100 dollar."
pattern = r"\b([4-9]|[1-4][0-9]|5[0-6])\b"

matches = re.findall(pattern, sentence)
print(matches)
```

#####  Explanation to the results above
* \b asserts a word boundary to ensure the match is a standalone word.
* [4-9] matches any single digit between 4 and 9.
* [1-4][0-9] matches any two-digit number starting with 1-4 followed by any digit.
* 5[0-6] matches the numbers 50-56.
* \b asserts another word boundary to ensure the match is a standalone word.
